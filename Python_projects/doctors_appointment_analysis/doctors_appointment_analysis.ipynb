```python
plt.figure(figsize=(9, 4))
plt.hist(ctf['Доля КТ от требуемого количества'],
         range=(0, 5),
         bins=50
         )
plt.title('Распределения доли КТ от требуемого количества')
plt.xlabel('Доля КТ от требуемого количества')
plt.ylabel('Количество специальностей')
plt.xticks(np.arange(0, 5.5, 0.5))
plt.grid(alpha=0.5)
plt.show()
```

```python
ctf_10 = ctf.groupby('Наименование юр. лица')['Число врачей без расписания'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(9,4))
plt.pie(ctf_10, labels=ctf_10.index, autopct='%1.1f%%')
plt.title('Распределением количества врачей без расписания по медицинским организациям (топ-10)')
plt.show()
```

```python
ctf_ct = ctf.groupby('Должность')['Количество КТ доступных для онлайн записи'].sum().sort_values(ascending=False).head()
```

```python
plt.figure(figsize=(15,5))
plt.bar(pd.Series(ctf_ct).index.str.replace('врач-', ''), ctf_ct.values)
plt.title('Рейтинг специальностей по количеству талонов для онлайн записи')
plt.show()
```

```python
ctf_fraction = ctf.groupby('Должность')['Доля КТ от требуемого количества'].mean().sort_values(ascending=False)
ctf_fraction
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=ctf_fraction)
plt.show()
```
