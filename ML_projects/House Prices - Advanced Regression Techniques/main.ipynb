```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
def label_encoder(df):
    df_encoded = df.copy()
    for column in df_encoded.columns:
        if df_encoded[column].dtype == 'object':
            df_encoded[column] = df_encoded[column].astype(str)
            df_encoded[column] = le.fit_transform(df_encoded[column])
    return df_encoded
train_df_le = label_encoder(train_df)
```

```python
train_df_le.head()
```

```python
# Избавляемся от NaN
train_df_le['LotFrontage'] = train_df_le['LotFrontage'].fillna(train_df_le['LotFrontage'].median())
train_df_le[['GarageYrBlt', 'MasVnrArea']] = train_df_le[['GarageYrBlt', 'MasVnrArea']].fillna(0)
```

```python
# Проверяем на NaN 
train_df_le.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Разделение данных
X = train_df_le.drop('SalePrice', axis=1)
Y = train_df_le['SalePrice']
```

```python
# Разделяем на тренировочую выборку и валидирующую 
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train.isnull().sum()
```

```python
# Пробуем разные модели 
models = {
    'RandomForestRegressor': RandomForestRegressor(n_estimators=100, random_state=42),
    'GradientBoostingRegressor': GradientBoostingRegressor (n_estimators=100, random_state=42)
}
```

```python
for name, model in models.items():
    model.fit(X_train, Y_train) # Обучение модели
    predictions = model.predict(X_val) # Предсказание на валидационных данных
    # Метрики регрессии
    mse = mean_squared_error(Y_val, predictions) 
    rmse = np.sqrt(mse)
    r2 = r2_score(Y_val, predictions)
    print(f'{name} RMSE: {rmse:.4f}, R^2: {r2:.4f}') 
```

```python
# Смотрим важность признаков 
feature_importance = pd.DataFrame({
    'feature' : X.columns,
    'importance' : model.feature_importances_
}).sort_values('importance', ascending=False)
feature_importance
```

```python
# Подготавливаем тестовые данные 
test_df_le = label_encoder(test_df)
test_df_le['LotFrontage'] = test_df_le['LotFrontage'].fillna(test_df_le['LotFrontage'].median())
test_df_le[['GarageYrBlt', 'MasVnrArea','BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 'GarageCars', 'GarageArea']] = test_df_le[['GarageYrBlt', 'MasVnrArea','BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 'GarageCars', 'GarageArea']].fillna(0)
```

```python
#  Обучаем модель на всех трен данных
final_model = GradientBoostingRegressor(n_estimators=100, random_state=42) 
final_model.fit(X, Y)
```

```python
# Предсказание для тестовых данных
test_predictions = final_model.predict(test_df_le)
```

```python
submission = pd.DataFrame({
    'Id': test_df_le['Id'],
    'SalePrice': test_predictions
})
```
