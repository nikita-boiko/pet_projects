```python
def minimum_total(triangle):
    n = len(triangle)
    prev = triangle[0][:] 
    
    for i in range(1, n):
        curr = [0] * len(triangle[i])
        for j in range(len(triangle[i])):
            if j == 0:
                curr[j] = prev[j] + triangle[i][j]
            elif j == len(triangle[i]) - 1:
                curr[j] = prev[j-1] + triangle[i][j]
            else:
                curr[j] = min(prev[j-1], prev[j]) + triangle[i][j]
        prev = curr  
    
    return min(prev)
```
